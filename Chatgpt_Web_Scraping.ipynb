```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time
import random

app_id = "com.openai.chatgpt"  # ChatGPT 
target_total = 100000          # Target size
batch_size = 200               
save_every = 20000             

all_reviews = []
continuation_token = None

while len(all_reviews) < target_total:
    result, continuation_token = reviews(
        app_id,
        lang='en',         
        country='us',
        sort=Sort.NEWEST,  
        count=batch_size,
        continuation_token=continuation_token
    )

    if not result:
        print("No more reviews")
        break

    all_reviews.extend(result)
    print(f"Have saved: {len(all_reviews)} reviews")

    if len(all_reviews) % save_every == 0:
        df = pd.DataFrame(all_reviews)
        df.to_csv(f"chatgpt_reviews_partial_{len(all_reviews)}.csv", index=False)
        print(f"Save file in progress: chatgpt_reviews_partial_{len(all_reviews)}.csv")

    time.sleep(random.uniform(1.2, 2.0))

df = pd.DataFrame(all_reviews)
df.to_csv(f"chatgpt_reviews_final_{len(all_reviews)}.csv", index=False)
print(f"In total, {len(all_reviews)} reviews have been downloaded.")
```
